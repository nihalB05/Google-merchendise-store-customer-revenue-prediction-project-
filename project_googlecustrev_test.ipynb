```python
placeholder_values = [
    "not available in demo dataset", "(not set)", "", " : ", "unknown.unknown", "NaN", "null", "None"
]
df_test.replace(placeholder_values, np.nan, inplace=True)
```

```python
nulled_columns = [
    'device_mobileInputSelector', 'device_mobileDeviceBranding',
    'geoNetwork_cityId', 'geoNetwork_networkLocation',
    'device_screenResolution', 'device_screenColors', 'device_language',
    'device_flashVersion', 'device_mobileDeviceMarketingName',
    'device_mobileDeviceInfo', 'geoNetwork_latitude',
    'device_mobileDeviceModel', 'geoNetwork_longitude',
    'trafficSource_adwordsClickInfo.criteriaParameters',
    'device_operatingSystemVersion', 'device_browserSize',
    'device_browserVersion'
]
for col in nulled_columns:
    if col in df_test.columns:
        df_test.drop(col, axis=1, inplace=True)

# Fill missing values (same as training)
categorical_columns = [
    "device_operatingSystem", "geoNetwork_continent",
    "geoNetwork_subContinent", "geoNetwork_country",
    "trafficSource_source", "trafficSource_medium"
]
df_test[categorical_columns] = df_test[categorical_columns].fillna("unknown")

numeric_columns = ['totals_pageviews', 'totals_hits', 'totals_visits', 'totals_transactions']
for col in numeric_columns:
    df_test[col] = df_test[col].fillna(df_test[col].median())
```

```python
# Date features
df_test['date'] = pd.to_datetime(df_test['date'], format='%Y%m%d')
df_test['visit_day'] = df_test['date'].dt.day
df_test['visit_month'] = df_test['date'].dt.month
df_test['visit_year'] = df_test['date'].dt.year

# Time features from visitStartTime
df_test['visitStartTime'] = pd.to_datetime(df_test['visitStartTime'], unit='s')
df_test['visit_hour'] = df_test['visitStartTime'].dt.hour
df_test['visit_day_of_week'] = df_test['visitStartTime'].dt.dayofweek  # Monday=0
df_test['is_weekend'] = df_test['visit_day_of_week'].isin([5, 6]).astype(int)

# Categorical grouping (geoNetwork_country)
country_counts = df_test["geoNetwork_country"].value_counts()
df_test['geoNetwork_country'] = df_test['geoNetwork_country'].apply(
    lambda x: x if country_counts.get(x, 0) > 100 else 'Other'
)

# Season and month flags
df_test['season'] = df_test['visit_month'].apply(
    lambda x: 'Winter' if x in [12, 1, 2] else
              'Spring' if x in [3, 4, 5] else
              'Summer' if x in [6, 7, 8] else
              'Autumn'
)
df_test['is_month_end'] = df_test['visit_day'].apply(lambda x: 1 if x >= 26 else 0)
df_test['is_month_start'] = df_test['visit_day'].apply(lambda x: 1 if x <= 5 else 0)

df_test['paying_cust']=(df_test["totals_transactionRevenue"]>0).astype(int)

# Binary features
binary_cols = ['device_isMobile', 'paying_cust']
for col in binary_cols:
    df_test[col] = df_test[col].astype(int)
```

```python
columns_to_drop = [
    "date", "visitId", "day_of_week", "visitStartTime",
    "fullVisitorId", "totals_totalTransactionRevenue"
]
df_test = df_test.drop(columns=[col for col in columns_to_drop if col in df_test.columns])
```

```python
extra_columns = [
    'geoNetwork_region', 'geoNetwork_metro', 'geoNetwork_city',
    'geoNetwork_networkDomain', 'totals_timeOnSite', 'totals_sessionQualityDim',
    'totals_newVisits', 'totals_bounces', 'trafficSource_referralPath',
    'trafficSource_campaign', 'trafficSource_keyword', 'trafficSource_adContent',
    'trafficSource_isTrueDirect', 'trafficSource_adwordsClickInfo.page',
    'trafficSource_adwordsClickInfo.slot', 'trafficSource_adwordsClickInfo.gclId',
    'trafficSource_adwordsClickInfo.adNetworkType', 'trafficSource_adwordsClickInfo.isVideoAd',"paying_cust"
]

df_test = df_test.drop(columns=extra_columns)
```

```python
import pickle

# Load the preprocessor (or use the one from your notebook)
with open('/content/best_model_final1.pkl', 'rb') as f:
    best_model = pickle.load(f)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
```

```python
pred=best_model.predict(df_test)
```

```python
df_test_visitid=df_test_visitid.to_frame()

print(df_test_visitid.columns)
```

```python
submission = pd.DataFrame({
    'fullVisitorId': df_test_visitid['fullVisitorId'].astype(str),  # Ensure this column is present
    'PredictedLogRevenue': pred
})
```

```python
submission.head()
```

```python
# Step 1: Aggregate predictions by fullVisitorId
submission_agg = submission.groupby('fullVisitorId')['PredictedLogRevenue'].sum().reset_index()

# Step 2: Identify missing fullVisitorId values
expected_ids = set(df_test_visitid['fullVisitorId'].unique())
missing_ids = expected_ids - set(submission_agg['fullVisitorId'])

# Step 3: Add missing fullVisitorId values with 0 revenue
if missing_ids:
    print(f"Adding {len(missing_ids)} missing fullVisitorId values with 0 revenue")
    missing_df = pd.DataFrame({'fullVisitorId': list(missing_ids), 'PredictedLogRevenue': 0})
    submission_agg = pd.concat([submission_agg, missing_df], ignore_index=True)

# Step 4: Verify final row count
print(f"Final submission has {len(submission_agg)} rows")

```

```python
# Check for duplicates in the aggregated submission
duplicates = submission_agg[submission_agg.duplicated('fullVisitorId', keep=False)]
print(f"Duplicate fullVisitorId values: {len(duplicates)}")
```
