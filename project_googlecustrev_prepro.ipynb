```python

def parse_single_json(df, col):
    """Parse columns with a single dictionary."""
    expanded = pd.json_normalize(df[col].apply(json.loads))
    expanded.columns = [f"{col}_{c}" for c in expanded.columns]
    return df.join(expanded).drop(col, axis=1)

# Parse single-dictionary columns
df_train = parse_single_json(df_train, "device")
df_train = parse_single_json(df_train, "geoNetwork")
df_train = parse_single_json(df_train, "totals")
df_train = parse_single_json(df_train, "trafficSource")
```

```python
def list_json_parse(df,col):
  #df[col]=df[col].apply(json.loads)
  exploded=df[col].explode().reset_index()
  expanded=pd.json_normalize(exploded[col])
  expanded.columns=[f"{col}_{c}" for c in expanded.columns]
  if f"{col}_index" in expanded.columns and f"{col}_value" in expanded.columns:
    pivoted = expanded.pivot(
            index="index",
            columns=f"{col}_index",
            values=f"{col}_value"
        )

    pivoted.columns = [f"{col}_{c}" for c in pivoted.columns]

  else:
    pivoted=expanded
  df=df.join(pivoted,how="left").drop(col,axis=1)
  return df
```

```python
def is_valid_json(value):
    try:
        json.loads(value)
        return True
    except:
        return False

# Check for invalid JSON rows
invalid_rows = df_train[~df_train["hits"].apply(is_valid_json)]
print(f"Number of invalid rows: {len(invalid_rows)}")
print(invalid_rows["hits"].head())
```

```python
df_train=list_json_parse(df_train,"hits")
df_train=list_json_parse(df_train,"customDimensions")
```
