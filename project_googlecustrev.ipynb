```python
import numpy as np
```

```python
df_train.replace('(not provided)', np.nan, inplace=True)
```

```python
threshold = 0.9  # Drop columns with >90% missing values
df_train = df_train.dropna(axis=1, thresh=int(threshold * len(df_train)))
```

```python
categorical_columns = [
    "device_operatingSystem",
    "geoNetwork_continent", "geoNetwork_subContinent", "geoNetwork_country",
    "trafficSource_source", "trafficSource_medium"
]
df_train[categorical_columns] = df_train[categorical_columns].fillna("unknown")
```
